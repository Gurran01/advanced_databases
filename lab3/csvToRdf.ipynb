```python
import csv
import pandas as pd
```

```python
assigned_hours_df = pd.read_csv("csv_files/Assigned_Hours.csv", skipinitialspace=True)
course_instances_df = pd.read_csv("csv_files/Course_Instances.csv", skipinitialspace=True)
course_plannings_df = pd.read_csv("csv_files/Course_plannings.csv", skipinitialspace=True)
courses_df = pd.read_csv("csv_files/Courses.csv", skipinitialspace=True)
programme_courses_df = pd.read_csv("csv_files/Programme_Courses.csv", skipinitialspace=True)
programmes_df = pd.read_csv("csv_files/Programmes.csv", skipinitialspace=True)
registrations_df = pd.read_csv("csv_files/Registrations.csv", skipinitialspace=True)
reported_hours_df = pd.read_csv("csv_files/Reported_Hours.csv", skipinitialspace=True)
senior_teachers_df = pd.read_csv("csv_files/Senior_Teachers.csv", skipinitialspace=True)
students_df = pd.read_csv("csv_files/Students.csv", skipinitialspace=True)
teaching_assistants_df = pd.read_csv("csv_files/Teaching_Assistants.csv", skipinitialspace=True)

teachers_df = pd.concat([teaching_assistants_df, senior_teachers_df], ignore_index=True)
```

```python
# Output RDF Turtle format for Courses
with open("ttl_files/courses.ttl", "w") as f:
    # Write prefixes
    f.write("@prefix : <http://example.org/university#> .\n")
    f.write("@prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> .\n")
    f.write("@prefix xsd: <http://www.w3.org/2001/XMLSchema#> .\n\n")

    merged_df = pd.merge(courses_df, programme_courses_df,
                         left_on="Course code", right_on="Course", how="left")
    # Loop over each row
    for index, row in merged_df.iterrows():
        course_uri = f":course{row['Course code']}"
        f.write(f"{course_uri} rdf:type :Course ;\n")
        f.write(f"    :courseCode \"{row['Course code']}\"^^xsd:integer ;\n")
        f.write(f"    :courseName \"{row['Course name']}\"^^xsd:string ;\n")
        f.write(f"    :credit \"{row['Credits']}\"^^xsd:decimal ;\n")
        f.write(f"    :level \"{row['Level']}\"^^xsd:string ;\n")  
        f.write(f"    :courseType \"{row['Course Type']}\"^^xsd:string .\n\n")
```

```python
# Output RDF Turtle format for ProgrammeCourses
with open("ttl_files/programmeCourses.ttl", "w") as f:
    # Write prefixes
    f.write("@prefix : <http://example.org/university#> .\n")
    f.write("@prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> .\n")
    f.write("@prefix xsd: <http://www.w3.org/2001/XMLSchema#> .\n\n")

    # Loop over each row
    for index, row in programme_courses_df.iterrows():
        academic_year = row['Academic Year']
        study_year = str(row['Study Year'])
        course_code = row['Course']
        programme_code = row['Programme code']

        pc_uri = f":programmeCourse_{academic_year}_{study_year}_{course_code}_{programme_code}"
        
        f.write(f"{pc_uri} rdf:type :ProgrammeCourses ;\n")
        f.write(f"    :pCAcademicYear \"{row['Academic Year']}\"^^xsd:string ;\n")
        f.write(f"    :pCStudyYear \"{row['Study Year']}\"^^xsd:decimal ;\n")
        f.write(f"    :course :course{course_code} ;\n")
        f.write(f"    :programme :programme{programme_code} .\n\n")
```

```python
# Output RDF Turtle format for Programmes
with open("ttl_files/programmes.ttl", "w") as f:
    # Write prefixes
    f.write("@prefix : <http://example.org/university#> .\n")
    f.write("@prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> .\n")
    f.write("@prefix xsd: <http://www.w3.org/2001/XMLSchema#> .\n\n")

    merged_df = pd.merge(programmes_df, courses_df,
                         left_on="Department name", right_on="Department", how="left")
    # Loop over each row
    for index, row in merged_df.iterrows():
        programme_uri = f":programme{row['Programme code']}"
        f.write(f"{programme_uri} rdf:type :Programme ;\n")
        f.write(f"    :programmeCode \"{row['Programme code']}\"^^xsd:integer ;\n")
        f.write(f"    :programmeName \"{row['Programme name']}\"^^xsd:string ;\n")
        f.write(f"    :director :seniorTeacher{row['Director']} ;\n")
        f.write(f"    :programDivision :division{row['Department']} .\n\n") 
```

```python
# Output RDF Turtle format for Students
with open("ttl_files/students.ttl", "w") as f:
    # Write prefixes
    f.write("@prefix : <http://example.org/university#> .\n")
    f.write("@prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> .\n")
    f.write("@prefix xsd: <http://www.w3.org/2001/XMLSchema#> .\n\n")
    
    # Assume 'graduated' is a column in your DataFrame
    students_df['Graduated'] = students_df['Graduated'].astype(str).str.lower()

    # Loop over each row
    for index, row in students_df.iterrows():
        student_uri = f":student{row['Student id']}"
        f.write(f"{student_uri} rdf:type :Student ;\n")
        f.write(f"    :studentId \"{row['Student id']}\"^^xsd:string ;\n")
        f.write(f"    :studentName \"{row['Student name']}\"^^xsd:string ;\n")
        f.write(f"    :graduated \"{row['Graduated']}\"^^xsd:boolean ;\n") # needs improvement should be an integer
        f.write(f"    :studentEnrollmentYear \"{row['Year']}\"^^xsd:integer ;\n")
        f.write(f"    :enrolledIn :programme{row['Programme']} .\n\n") 
```

```python
# Output RDF Turtle format for course instances

with open("ttl_files/courseInstance.ttl", "w") as f:
    # Write prefixes
    f.write("@prefix : <http://example.org/university#> .\n")
    f.write("@prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> .\n")
    f.write("@prefix xsd: <http://www.w3.org/2001/XMLSchema#> .\n\n")

    merged_df = pd.merge(course_instances_df, course_plannings_df,
                     left_on="Instance_id", right_on="Course", how="left")
    # Loop over each row
    for index, row in merged_df.iterrows():
        academic_year = row['Academic year']
        study_period = str(row['Study period'])
        course_code = row['Course code']
        courseInstance_uri = f":courseInstance_{academic_year}_{study_period}_{course_code}"

        f.write(f"{courseInstance_uri} rdf:type :CourseInstance ;\n")
        f.write(f"    :cIAcademicYear \"{row['Academic year']}\"^^xsd:string ;\n")
        f.write(f"    :cIStudyPeriod \"{row['Study period']}\"^^xsd:decimal ;\n")
        f.write(f"    :planningNumStudents :{row['Planned number of Students']}\"^^xsd:integer ;\n")
        f.write(f"    :course :course{course_code} .\n\n")
```

```python
department_csv_files = [courses_df, programmes_df, senior_teachers_df, teaching_assistants_df]
possible_column_names = ["Department", "Department name"]
departments = set()

# Loop through each file
for file in department_csv_files:
    for col in possible_column_names:
        if col in file.columns:
            departments.update(file[col].dropna().unique())

# Create a DataFrame and save to CSV
departments_df = pd.DataFrame(sorted(departments), columns=["Department"])
departments_df.to_csv("csv_files/department.csv", index=False)

departments_name_df = pd.read_csv("csv_files/department.csv", skipinitialspace=True)

# Output RDF Turtle format for department
with open("ttl_files/department.ttl", "w") as f:
    # Write prefixes
    f.write("@prefix : <http://example.org/university#> .\n")
    f.write("@prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> .\n")
    f.write("@prefix xsd: <http://www.w3.org/2001/XMLSchema#> .\n\n")

    # Loop over each row
    for index, row in departments_name_df.iterrows():

        department_uri = f":department{row['Department']}"
        f.write(f"{department_uri} rdf:type :Department ;\n")
        f.write(f"    :departmentName \"{row['Department']}\"^^xsd:string .\n\n") 
```

```python
# Output RDF Turtle format for division
division_csv_files = [courses_df, senior_teachers_df, teaching_assistants_df]
possible_column_names = ["Division", "Division name"]
divisions = set()

# Loop through each file
for file in division_csv_files:
    for col in possible_column_names:
        if col in file.columns:
            divisions.update(file[col].dropna().unique())

# Create a DataFrame and save to CSV
divisions_df = pd.DataFrame(sorted(divisions), columns=["Division"])
divisions_df.to_csv("csv_files/division.csv", index=False)

divisions_name_df = pd.read_csv("csv_files/division.csv", skipinitialspace=True)

with open("ttl_files/division.ttl", "w") as f:
    # Write prefixes
    f.write("@prefix : <http://example.org/university#> .\n")
    f.write("@prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> .\n")
    f.write("@prefix xsd: <http://www.w3.org/2001/XMLSchema#> .\n\n")

    # Loop over each row
    for index, row in divisions_name_df.iterrows():

        division_uri = f":division{row['Division']}"
        f.write(f"{division_uri} rdf:type :Division ;\n")
        f.write(f"    :divisionName \"{row['Division']}\"^^xsd:string .\n\n") 
```

```python
# Output RDF Turtle format for divisionInDepartment
division_csv_files = [courses_df, senior_teachers_df, teaching_assistants_df]
possible_column_names = ["Division", "Division name"]
divisionsInDepartment = set()

# Loop through each file
for file in division_csv_files:
    for col in possible_column_names:
        if col in file.columns:
            divisionsInDepartment.update(file[col].dropna().unique())

# Create a DataFrame and save to CSV
divisionsInDepartment_df = pd.DataFrame(sorted(divisionsInDepartment), columns=["Division"])
divisionsInDepartment_df.insert(
    0,  # Insert at column index 0 (leftmost)
    "Department", 
    divisionsInDepartment_df["Division"].str.split('-').str[0]
)

divisionsInDepartment_df.to_csv("csv_files/divisionsInDepartment.csv", index=False)

# Read back for consistency
divisionsInDepartment_name_df = pd.read_csv("csv_files/divisionsInDepartment.csv", skipinitialspace=True)

# Write RDF Turtle
with open("ttl_files/divisionsInDepartment.ttl", "w") as f:
    # Write prefixes
    f.write("@prefix : <http://example.org/university#> .\n")
    f.write("@prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> .\n")
    f.write("@prefix xsd: <http://www.w3.org/2001/XMLSchema#> .\n\n")

    # Loop over each row
    for index, row in divisionsInDepartment_name_df.iterrows():
        division_uri = f":division{row['Division']}"
        department_uri = f":department{row['Department']}"

        f.write(f"{division_uri} rdf:type :DivisionsInDepartment ;\n")
        f.write(f"    :divisionName \"{row['Division']}\"^^xsd:string ;\n")
        f.write(f"    :DepartmentName {department_uri} .\n\n")
```

```python
# Standardize column names

# assigned_hours_df = pd.read_csv("csv_files/Assigned_Hours.csv", skipinitialspace=True)
# reported_hours_df = pd.read_csv("csv_files/Reported_Hours.csv", skipinitialspace=True)

# print("Reported columns:", reported_hours_df.columns.tolist())

assigned_hours_df.rename(columns={
    "Hours": "Assigned Hours"
}, inplace=True)

reported_hours_df.rename(columns={
    "Course code": "Course Instance",  # Aligning with Assigned_Hours
    "Hours": "Reported Hours"
}, inplace=True)

# Ensure merge columns are strings
assigned_hours_df["Course Instance"] = assigned_hours_df["Course Instance"].astype(str)
assigned_hours_df["Teacher Id"] = assigned_hours_df["Teacher Id"].astype(str)
reported_hours_df["Course Instance"] = reported_hours_df["Course Instance"].astype(str)
reported_hours_df["Teacher Id"] = reported_hours_df["Teacher Id"].astype(str)

# Merge
merged_df = pd.merge(
    assigned_hours_df,
    reported_hours_df,
    on=["Course Instance", "Teacher Id"],
    how="outer"
)
merged_df
# Select and reorder columns
courseInstanceStaff_dfb = merged_df[["Course Instance", "Teacher Id", "Assigned Hours", "Reported Hours"]]

# Save
courseInstanceStaff_dfb.to_csv("csv_files/courseInstanceStaff.csv", index=False)

courseInstanceStaff_df = pd.read_csv("csv_files/courseInstanceStaff.csv", skipinitialspace=True)


# Output RDF Turtle format for CourseDivision
with open("ttl_files/courseInstanceStaff.ttl", "w") as f:
    # Write prefixes
    f.write("@prefix : <http://example.org/university#> .\n")
    f.write("@prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> .\n")
    f.write("@prefix xsd: <http://www.w3.org/2001/XMLSchema#> .\n\n")

    # Loop over each row
    for index, row in courseInstanceStaff_df.iterrows():
        courseInstance = row["Course Instance"]
        teacherId = row["Teacher Id"]
        courseInstanceStaff_uri = f":courseInstanceStaff_{courseInstance}_{teacherId}"
        f.write(f"{courseInstanceStaff_uri} rdf:type :courseInstanceStaff ;\n")
        f.write(f"    :courseInstance :courseInstance\"{row['Course Instance']}\"^^xsd:string ;\n")
        f.write(f"    :teacherId :teacher\"{row['Teacher Id']}\"^^xsd:string ;\n")
        f.write(f"    :assignedHours\"{row['Assigned Hours']}\"^^xsd:string ;\n")
        f.write(f"    :reportedHours\"{row['Reported Hours']}\"^^xsd:string .\n\n")        
```
